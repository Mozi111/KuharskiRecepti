```python
# naložimo paket
import pandas as pd

# naložimo razpredelnico, s katero bomo delali
recepti = pd.read_csv('recepti.csv', index_col='id')

# ker bomo delali z velikimi razpredelnicami, povemo, da naj se vedno izpiše le 20 vrstic
pd.options.display.max_rows = 20
```

```python
recepti
```

```python
recepti_po_kategorijah = recepti.groupby('kategorija')
recepti_po_kategorijah.size().sort_values()
```

```python
recepti_po_spolih = recepti.groupby('spol')
recepti_po_spolih.size().sort_values()
```

```python
recepti_po_spolih['stmnenj'].mean().sort_values()
```

```python
recepti_po_spolih['stfoto'].mean().sort_values()
```

```python
recepti_po_avtorjih = recepti.groupby('avtor')
recepti_po_avtorjih.size().sort_values()
```

```python
recepti_po_casu = recepti.groupby('priprava')
recepti_po_casu.size().sort_values()
```

```python
recepti_po_kolicinah = recepti.groupby('kolicina')
recepti_po_kolicinah.size().sort_values()
```
